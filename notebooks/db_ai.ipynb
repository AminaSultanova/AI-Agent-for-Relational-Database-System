````python
import google.generativeai as genai
import mysql.connector
import pandas as pd

genai.configure(api_key="AIzaSyCvZvXszPNu1Kx5voMNmXae8qRbRD7JbBc")

model = genai.GenerativeModel("models/gemini-2.5-flash")
````

````python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="78900987",
    database="ecommerce"
)

cursor = conn.cursor()
````

````python
def generate_sql(question: str) -> str:
    prompt = f"""
    You are an expert MySQL assistant.
    Write ONLY a MySQL SELECT query that answers this question:

    {question}

    Important rules:
    - Only use tables: customers, orders, order_items, products, sellers, order_payments, order_reviews, geolocation.
    - Do NOT use PRAGMA or sqlite_master.
    - Do NOT explain the query.
    - Output only SQL.
    """

    response = model.generate_content(prompt)
    sql = response.text.strip().strip("```sql").strip("```")
    return sql
````

````python
def run_sql(sql):
    cursor.execute(sql)
    rows = cursor.fetchall()
    cols = [col[0] for col in cursor.description]
    return pd.DataFrame(rows, columns=cols)
````

````python
question = "How many rows are in the customers table?"
sql = generate_sql(question)

print("Generated SQL:\n", sql)

df = run_sql(sql)
df
````

````python
question = """Find the customer states with the highest average delivery delay.
Delivery delay = order_delivered_customer_date - order_estimated_delivery_date.
Return the top 10 states with the biggest delays.
"""

sql = generate_sql(question)
print(sql)

df = run_sql(sql)
df
````

````python
question = """Find the product categories that generate the highest total revenue.
Revenue = price  order_item quantity (order_items.product_price  order_items.order_item_id?).
Use tables: products, order_items. 
Join: order_items.product_id = products.product_id.
Return top 10 categories by total revenue."""

sql = generate_sql(question)

print("Generated SQL:\n", sql)

df = run_sql(sql)
df
````

````python
question = """
Find sellers with consistently low review scores.
Join order_reviews -> orders -> order_items -> sellers.
Calculate AVG(review_score) for each seller.
Return sellers with avg_score < 3.
Sort by avg_score ascending.
Show seller_id, avg_score, number_of_reviews.
"""

sql = generate_sql(question)
print(sql)

df = run_sql(sql)
df
````

````python
question = """
Identify high-value customers based on:
1) Total spending (SUM(order_items.price))
2) Number of orders.

Use tables: customers, orders, order_items.
Join: customers.customer_id = orders.customer_id,
      orders.order_id = order_items.order_id.

Return top 20 customers sorted by total_spent (DESC), then order_count (DESC).
Show: customer_id, customer_city, customer_state, order_count, total_spent.
"""


sql = generate_sql(question)
print(sql)

df = run_sql(sql)
df
````
